```python
from matplotlib import pyplot as plt
```

```python
x=["roshan","raju","shambu","shiva","richu","kithu","ravi","shiju","shaaji","vichu","hdcdf","ughdc","hisdgc","hdcg"]
y=[5,6,9,8,4,5,5,6,8,8,7,8,8,1]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,marker="o",color="g",linestyle="--",lw=3)
plt.title("marklist")
plt.xlabel("student_name")
plt.ylabel("mark")
plt.grid()
```

```python
x=["roshan","raju","shambu","shiva","richu","kithu","ravi","shiju","shaaji","vichu","hdcdf","ughdc","hisdgc","hdcg"]
y=[5,6,9,8,4,5,5,6,8,8,7,8,8,1]
z=[6,8,10,5,6,7,7,6,7,8,7,8,9,7]
```

```python
plt.plot(x,y,marker="o",color="r",linestyle="--",lw=2,label="halfyear_exam")
plt.plot(x,z,marker="o",color="g",linestyle=":",lw=1,label="finalyear_exam")
plt.xlabel=("student_name")
plt.ylabel=("marklist")
plt.legend(loc="upper right")
```
