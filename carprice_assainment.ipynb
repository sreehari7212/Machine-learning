```python
dummie1=pd.get_dummies(df["CarName"])
dummie2=pd.get_dummies(df["fueltype"])
dummie3=pd.get_dummies(df["aspiration"])
dummie4=pd.get_dummies(df["carbody"])
dummie5=pd.get_dummies(df["drivewheel"])
dummie6=pd.get_dummies(df["enginetype"])
dummie7=pd.get_dummies(df["fuelsystem"])                        
```

```python
x=df.drop(["price"],axis=1)
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.25,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
